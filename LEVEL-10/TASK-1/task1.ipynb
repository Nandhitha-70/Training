```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("titanic.csv")
df.head()
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode(numeric_only=True).iloc[0])
df.describe()
categorical_features = ['sex', 'pclass', 'embarked', 'sibsp', 'parch', 'alone']

for feature in categorical_features:
    plt.figure(figsize=(7, 4))
    sns.countplot(x=feature, hue='survived', data=df)
    plt.title(f'Survival Count by {feature.capitalize()}')
    plt.show()
pd.crosstab(df['pclass'], df['survived'], normalize='index') * 100
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df.drop(columns=['deck', 'embark_town'], inplace=True)
```
